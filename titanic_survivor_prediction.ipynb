```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

# Preparing training data 

Reading train Data

```python
df = pd.read_csv('train.csv', index_col='PassengerId')
df
```

```python
y_train = df.iloc[:,[0]]
y_train
```

creating training data set

```python
x_train = df.loc[:,['Pclass','Sex', 'Age','SibSp']]
x_train
```

```python
mean = x_train['Age'].mean()
x_train['Age'] = x_train['Age'].fillna(mean)
```

```python
x_train['Age'] = x_train['Age'].astype(int)
x_train
```

```python
x_train['sex_male'] = x_train['Sex'].apply(lambda x: 1 if x == 'male' else 0)

x_train['sex_female'] = x_train['Sex'].apply(lambda x: 1 if x == 'female' else 0)
x_train
```

```python
x_train = x_train.loc[:, x_train.columns != 'Sex']
x_train
```

# Preparing test data


```python
test_df = pd.read_csv('test.csv',index_col='PassengerId')
test_df
```

```python
#Creating test data

x_test = test_df.loc[:,['Pclass','Sex', 'Age','SibSp']]
x_test
```

```python
#removing na values from Age column and filling with mean value

x_test['Age'] = x_test['Age'].fillna(x_test['Age'].mean())
x_test
```

```python
#converting age value into integer
x_test['Age'] = x_test['Age'].astype(int)
```

```python
x_test['sex_male'] = x_test['Sex'].apply(lambda x: 1 if x == 'male' else 0)

x_test['sex_female'] = x_test['Sex'].apply(lambda x: 1 if x == 'female' else 0)
x_test
```

```python
x_test = x_test.loc[:, x_test.columns != 'Sex']
x_test
```

# Creating model and Predicting

```python
logisticRegr = LogisticRegression()
```

```python
logisticRegr.fit(x_train, y_train)
```

# Predicting on test data

```python
y_pred = logisticRegr.predict(x_test)
y_pred
```

# Results 

```python
result_df = x_test.iloc[:,5]
result_df
```

```python
result = pd.read_csv('gender_submission.csv')
result
```

```python
from sklearn import metrics
acc = metrics.accuracy_score(result["Survived"], y_pred)
print(f'The accuracy of the classification of points from the test set is {acc * 100 :.1f}%.')

```
