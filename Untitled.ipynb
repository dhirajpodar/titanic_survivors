```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('train.csv', index_col='PassengerId')
df
```

```python
y_train = df.iloc[:,[0]]
y_train
```

```python
x_train = df.loc[:,['Pclass','Sex', 'Age','SibSp','Fare']]
x_train
```

```python
mean = x_train['Age'].mean()
x_train['Age'] = x_train['Age'].fillna(mean)
```

```python
x_train['Age'] = x_train['Age'].astype(int)
x_train
```

```python
x_train['Sex'].replace('male',1, inplace=True)
x_train['Sex'].replace('female',0, inplace=True)
x_train
```

```python
logisticRegr = LogisticRegression()
```

```python
logisticRegr.fit(x_train, y_train)
```

```python
test_df = pd.read_csv('test.csv',index_col='PassengerId')
test_df
```

```python
x_test = test_df.loc[:,['Pclass','Sex', 'Age','SibSp','Fare']]
x_test
```

```python
x_test['Age'] = x_test['Age'].fillna(x_test['Age'].mean())
x_test
```

```python
x_test['Age'] = x_test['Age'].astype(int)
```

```python
x_test['Sex'].replace('male',1, inplace=True)
x_test['Sex'].replace('female',0, inplace=True)
x_test
```

```python
x_test[x_test['Fare'].isnull()]
```

```python
x_test['Fare'] = x_test['Fare'].fillna(9)
```

```python
y_pred = logisticRegr.predict(x_test)
y_pred
```

```python
x_test['Survived'] = y_pred
x_test
```
